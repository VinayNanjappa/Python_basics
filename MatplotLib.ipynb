## MatplotLib Tutorial

Matplotlib is a plotting library for the python programming language and its numerical mathematical extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major pros of Matplotlib are:

1. Generally easy to get started with simple plots.
2. Support for custom lables and text.
3. Great control of every element in a figure.
4. High quality output in many formats.
5. Very customizable in general.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## Plotting using matplot lib
## Scattered plot

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.show()
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y-')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('X axis caption')
plt.ylabel('Y axis caption')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

plt.plot(x,y)
plt.show()
```

```python
# Compute the X and Y coordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1
# Set up the first such subplot has active
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'g--')
plt.title('Sine')

# Set the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r*')
plt.title('Cosine')

#Show the figure
plt.show()
```

```python
# Bar plot

x = [2,8,10]
y = [11,16,19]

x2 = [3,9,11]
y2 = [6,15,17]

plt.bar(x,y, align = 'center')
plt.bar(x2,y2, color = 'g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
plt.title("histogram")
plt.show()
```

## Box plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```

## Pie Chart

```python
# Data to plot
labels = 'Python','C++','Ruby','Java'
sizes = [215, 130, 245, 210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0,0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
