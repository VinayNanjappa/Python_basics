```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_iris = sns.load_dataset("iris")
```

```python
df_iris.head()
```

### Scattered plot in Matplot lib

```python
plt.scatter(df_iris['sepal_length'],df_iris['sepal_width'],c='r')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Length vs Width')
```

### Matplotlib - Lineplot

```python
plt.plot(df_iris['petal_length'],df_iris['petal_width'],'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Legth vs Width')
```

### Matplotlib subplot

```python
plt.subplot(1,2,1)
plt.plot(df_iris['sepal_length'],df_iris['sepal_width'],'r--')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.subplot(1,2,2)
plt.plot(df_iris['petal_length'],df_iris['petal_width'],'g*')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
plt.plot(df_iris['petal_length'],df_iris['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Legth vs Width')
```

### Matplot lib Bar plot

```python
plt.bar(df_iris['sepal_length'],df_iris['sepal_width'], align='center')
plt.bar(df_iris['petal_length'],df_iris['petal_width'],color='r')
plt.title('Length vs Width')
plt.xlabel('Length')
plt.ylabel('Width')
```

### Matplot lib - Histogram

```python
plt.hist(df_iris['sepal_length'],bins=20)
```

## Box plot - Matplotlib

```python
plt.boxplot(df_iris['petal_length'],vert=True,patch_artist=True)
```

## Join plot
Relationship between two numerical variable

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df_iris,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df_iris,kind='reg')
```

## Pair plot

A Pair plot is also known as a scattered plot in which one variable in the same data row is matched with another vartiable's value

```python
sns.pairplot(df_iris)
```

```python
# Classify the point in pair plot using hue
sns.pairplot(df_iris,hue='species')
```

```python
df_iris['species'].value_counts()
```

## Dist plot

Dist plot helps us to check the distribution of column features

```python
sns.distplot(df_iris['sepal_length'])
```

```python
sns.distplot(df_iris['sepal_length'],kde=False,bins=20)
```

## Categorical plots

```python
# count plots

sns.countplot('species',data=df_iris)
```

```python
# Bar plot

sns.barplot(x='sepal_length',y='species',data=df_iris)
```

```python
sns.boxplot(data=df_iris['sepal_length'],orient='v')
```

```python
sns.boxplot(x='sepal_length',y='species',data=df_iris,palette='rainbow')
```

```python
sns.boxplot(x='sepal_length',y='sepal_width',hue='species',data=df_iris)
```

## Violin plot

```python
sns.violinplot(x='species',y='sepal_length',data=df_iris)
```
