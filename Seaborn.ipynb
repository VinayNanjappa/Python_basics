```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

# Correlation with heat map

A correlation heatmap uses colored cells, typically in a monochromatic scale to show 2D correlation matrix (table) between 
two discrete dimensions or event types. It is very important for feature selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JoinPlot

A join plot allows you to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually scattered plot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot

A pari plot is also known as a scattered plot in which one variable in the same data row is matched with another varaible's value, like this: Pairs plot are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
# Classify the points in the pair plot using hue
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

## Dist plot

Dist plot helps us to check the ditribution of the column features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots

Seaborn helps us in doing analysis on catagorical data points. In this section we will discuss about.

1. boxplot
2. violinplot
3. countplot
4. barplot

```python
# Count plot

sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
# Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

##  Box plot

    A box and whisker plot (sometimes called a boxplot) is a graph that persents information from a five-number summary

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker", data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill",y="day",data=df,palette='rainbow')
```
